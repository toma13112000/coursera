## Foursquare API Data:

We will need data from different places in different communities in that particular area. To obtain this information, we will use "Foursquare" location information. Foursquare is a provider of location data, providing information about various places and events in the area of interest. This information includes place names, locations, menus, and even photos. Therefore, the foursquare positioning platform will be used as the only data source, because all required information required by the regulations can be obtained through the API. After finding the community listing, we connect to the Foursquare API to collect information about the locations in each community. For each community, we choose a radius of 100 meters. Data retrieved from Foursquare contains location information within a specified distance of the latitude and longitude of the zip code. The information obtained for each site is as follows:

    1. Neighborhood
    2. Neighborhood Latitude
    3. Neighborhood Longitude
    4. Venue
    5. Name of the venue e.g. the name of a store or restaurant
    6. Venue Latitude
    7. Venue Longitude
    8. Venue Category
